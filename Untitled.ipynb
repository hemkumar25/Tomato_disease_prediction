```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
```

```python
IMAGE_SIZE = 256
BATCH_SIZE=32
CHANNELS = 3
EPOCHS = 50
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "Tomato",
    shuffle = True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
class_names = dataset.class_names
class_names
```

```python
def get_dataset_partitions_tf(ds,train_split = 0.8 , val_split = 0.1, test_split=0.1, shuffle =True,shuffle_size = 10000):

    ds_size = len(ds)

    if shuffle:
        ds = ds.shuffle(shuffle_size,seed=24)

    train_size = int(train_split * ds_size)
    val_size = int(val_split*ds_size)

    train_ds = ds.take(train_size)

    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    
    return train_ds, val_ds, test_ds


train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

```python
rotation_augmentation = layers.RandomRotation(0.2)
flip_augmentation = layers.RandomFlip("horizontal_and_vertical")
```

```python
# data_augmentation = tf.keras.Sequential([
#     layers.RandomFlip("horizontal_and_vertical"),
#     layers.RandomRotation(0.2),
# ])

def visualize_augmentation(dataset,num_images):
    plt.figure(figsize=(15, 10))
    
    for image_batch, label_batch in dataset.take(1):
        # Apply data augmentation
        # augmented_images = data_augmentation(image_batch)
        
        for i in range(num_images):
            # Original images
            ax = plt.subplot(3, num_images, i + 1)
            plt.imshow(image_batch[i].numpy().astype("uint8"))
            plt.title(class_names[label_batch[i]])
            plt.axis("off")

            # Augmented images
            # ax = plt.subplot(3, num_images, i + 1 + num_images)
            # plt.imshow(augmented_images[i].numpy().astype("uint8"))
            # plt.title("Augmented")
            # plt.axis("off")

              # Augmented images from flipping
            flipped_image = flip_augmentation(image_batch[i:i+1])
            ax = plt.subplot(3, num_images, i + 1 + num_images)
            plt.imshow(flipped_image[0].numpy().astype("uint8"))
            plt.title("Flipped")
            plt.axis("off")

            # Augmented images from rotation
            rotated_image = rotation_augmentation(image_batch[i:i+1])
            ax = plt.subplot(3, num_images, i + 1 + 2*num_images)
            plt.imshow(rotated_image[0].numpy().astype("uint8"))
            plt.title("Rotated")
            plt.axis("off")
    
    plt.tight_layout()
    plt.show()

# Call the visualization function
visualize_augmentation(train_ds, num_images=4)
```
