```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
```

```python
IMAGE_SIZE = 256
BATCH_SIZE=32
CHANNELS = 3
EPOCHS = 50
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "Tomato",
    shuffle = True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
def get_dataset_partitions_tf(ds,train_split = 0.8 , val_split = 0.1, test_split=0.1, shuffle =True,shuffle_size = 10000):

    ds_size = len(ds)

    if shuffle:
        ds = ds.shuffle(shuffle_size,seed=24)

    train_size = int(train_split * ds_size)
    val_size = int(val_split*ds_size)

    train_ds = ds.take(train_size)

    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    
    return train_ds, val_ds, test_ds


train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)
```

```python
len(train_ds)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.Resizing(IMAGE_SIZE,IMAGE_SIZE),
    layers.Rescaling(1.0/255)
])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
])

for image_batch, label_batch in train_ds.take(1):
    original_image = image_batch[0]  # Get the first image
    original_label = label_batch[0]   # Get the corresponding label

# Apply data augmentation
augmented_image = data_augmentation(tf.expand_dims(original_image, axis=0))  # Add batch dimension

# Convert the augmented image to numpy for visualization
augmented_image = augmented_image.numpy()[0]  # Remove the batch dimension

# Plot the original and augmented images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image.numpy().astype("uint8") / 255.0)  # Normalize for display
plt.axis("off")

# Augmented Image
plt.subplot(1, 2, 2)
plt.title("Augmented Image")
plt.imshow(augmented_image)  # Already normalized
plt.axis("off")

plt.show()
```
